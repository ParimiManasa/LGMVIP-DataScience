### Importing the libraries

```python
import matplotlib.pyplot as plt
import cv2
```

### Reading the image

```python
image = cv2.imread("dog.jpeg")
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()
```

### Converting it to the gray scale

```python
grey_img=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey_img,cmap='gray')
plt.show()
```

### Inverting the grayscale image (Negetive)

```python
inverted_img=255-grey_img
```

```python
plt.imshow(inverted_img,cmap='coolwarm')
plt.show()
```

### Blurring the inverted grayscale image

```python
blur=cv2.GaussianBlur(inverted_img,(21,21),0)
```

```python
plt.imshow(blur,cmap='gray')
plt.show()
```

### Pencil sketch 

```python
inverted_blur=255-blur
pencilsketch = cv2.divide(grey_img,inverted_blur, scale=256.0)
```

```python
plt.imshow(pencilsketch, cmap='gray')
plt.show()
```

```python
#original image
plt.figure(figsize=(15,7))
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),cmap='gray')
plt.show()

# pencil sketch image
plt.figure(figsize=(15,7))
plt.imshow(pencilsketch,cmap='gray')
plt.show()
```
